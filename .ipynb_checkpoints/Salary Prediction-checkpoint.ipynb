## Data Splitting

```python
X = salaries_df.drop(columns=['salary_in_usd'])
Y = salaries_df['salary_in_usd']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
X_train.shape, y_train.shape
```

## Data Preprocessing

#### Column Transformer

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

ct = ColumnTransformer(transformers=[('ord_encoder', OrdinalEncoder(categories=[['FL','CT','PT','FT'], title_salary_df.index[::-1]]),
                                      ['employment_type', 'job_title']), 
                                     ('onehot', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), 
                                      ['experience_level','employee_residence', 'company_location', 'company_size'])], 
                       remainder='passthrough', sparse_threshold=0)

ct
```

## Model Selection

#### Models

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor, GradientBoostingRegressor, HistGradientBoostingRegressor

models = {
    'lin_reg': LinearRegression(),
    'knn': KNeighborsRegressor(),
    'svm': SVR(),
    'dec_tree': DecisionTreeRegressor(),
    'ran_forest': RandomForestRegressor(),
    'ex_trees': ExtraTreesRegressor(),
    'gbm': GradientBoostingRegressor(),
    'hgbm': HistGradientBoostingRegressor()
}
```

#### Cross Validations

```python
from sklearn.model_selection import cross_val_score

X_pr = ct.fit_transform(X)

cv_df = pd.DataFrame(index=models.keys(), columns=['CVs', 'R-squared'])

for i in models.keys():
    scores = cross_val_score(models[i], X_pr, Y, cv = 5)
    cv_df.loc[i, 'CVs'] = scores.round(2)
    cv_df.loc[i, 'R-squared'] = scores.mean()
    print(f'{i} is cross validated ...!')
```

```python
cv_df.sort_values(by = 'R-squared', ascending=False)
```

```python
from sklearn.pipeline import Pipeline

model = Pipeline(steps=[('ct', ct), ('gbm', GradientBoostingRegressor())])
model.fit(X_train, y_train)
```

## Model Evaluation

```python
model.score(X_test, y_test)
```

## HyperParameter Tuning

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'learning_rate' : [0.1, 0.2, 0.3, 0.05],
    'min_samples_split': [2, 3, 4, 5],
    'n_estimators': [50, 100, 150, 200],
    'min_samples_leaf': [1,2,3,4]
}

gs = GridSearchCV(GradientBoostingRegressor(), param_grid=param_grid, cv=5)
gs.fit(X_pr, Y)
```

```python
gs.best_score_
```

```python
gs.best_params_
```

## Model Re-Building

```python
model = Pipeline(steps=[('ct',ct),('gbm',GradientBoostingRegressor(learning_rate=0.05,min_samples_leaf=4,min_samples_split=3,n_estimators=100))])
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
